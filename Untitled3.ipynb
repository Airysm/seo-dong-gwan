```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터셋 생성
url = '/content/drive/MyDrive/Colab Notebooks/vgames2.csv'
df = pd.read_csv(url)
df.head()
```

```python
# 데이터 정보 확인
df.info()
```

```python
df.drop('Unnamed: 0', axis = 1, inplace = True)
# 결측치 값 확인
df.isnull().sum()
```

```python
# 결측치 바꾸고 중복 제거
df.dropna(how = 'any', inplace = True)
df.drop_duplicates(inplace = True)
```

```python
# 출고량에 있는 M, K값을 바꾸는 함수
def mk_change(num):
    if 'K' in num:
        num = num[:len(num)-1]
        num = float(num)*1000
    elif 'M' in num:
        num = num[:len(num)-1]
        num = float(num)*1000000
    else:
        num = float(num)*1000000
    return num

# 함수 적용
for column in ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']:
    df[column] = df[column].apply(mk_change)
```

```python
# Year에 있는 누락된 년도를 채우는 함수
def year_change(year):
    if len(str(int(year))) == 1:
        year = year+2000
    elif len(str(int(year))) == 2:
        if int(year) <= 21:
            year = year+2000
        else:
            year = year+1900
    return year

# 함수 적용
df['Year'] = df['Year'].apply(year_change)
```

```python
# Year 분포 확인
plt.hist(df['Year'])
plt.title('Year histogram')
plt.show()
```

**지역에 따라서 선호하는 게임 장르가 다를까** 라는 질문에 대답

```python
# NA_Sales 그래프
nasale = df.groupby('Genre')['NA_Sales'].mean()
plt.plot(nasale)
plt.title('NA')
plt.xticks(rotation = 45)
plt.show()
# NA에서 선호하는 게임 장르는 Platform, Shooter이다.
```

```python
# EU_Sales 그래프
eusale = df.groupby('Genre')['EU_Sales'].mean()
plt.plot(eusale)
plt.title('EU')
plt.xticks(rotation = 45)
plt.show()
# EU에서 선호하는 게임 장르는 Shooter, Platform, Racing이다.
```

```python
# JP_Sales 그래프
jpsale = df.groupby('Genre')['JP_Sales'].mean()
plt.plot(jpsale)
plt.title('JP')
plt.xticks(rotation = 45)
plt.show()
# JP에서 선호하는 게임 장르는 Role-Playing, Platform이다.
```

```python
# Other_Sales 그래프
othersale = df.groupby('Genre')['Other_Sales'].mean()
plt.plot(othersale)
plt.title('Other')
plt.xticks(rotation = 45)
plt.show()
# Other에서 선호하는 게임 장르는 Shooter, Racing, Platform, Sports이다.
```

```python
# 그래프를 보면 NA와 EU, Other의 출고량은 Platform, Shooter이 높은 값을 가진다.
# NA와 EU, Other의 출고량은 두 장르 말고도 모든 장르에서 유사한 패턴을 보인다.
# JP는 Platform장르에서 높은 값을 가지지만 Role-Playing가 높은 값을 가지면서
# 앞의 나라들과는 확연히 다르다.
# 결론적으로 일본을 제외한 전세계적으로는 선호하는 게임 장르가 유사하다.
```

**연도별 게임의 트렌드가 있을까** 라는 질문에 대답

```python
# 연도별 장르 그래프
df['total'] = df['NA_Sales'] + df['JP_Sales'] + df['EU_Sales'] + df['Other_Sales']

gen = list(df['Genre'].unique())

gen.sort()

fig, ax = plt.subplots(3,4, figsize=(25,13))

for i, ge in enumerate(gen):
    if i <= 3:
        r = 0
    elif i > 3 and i <= 7:
        r = 1
    elif i > 7 and i <= 11:
        r = 2

    j = i % 4

    df2 = df[df['Genre'] == ge]

    df3 = df2.groupby('Year')['total'].sum()

    ax[r,j].plot(df3)
    ax[r,j].set_title(ge)
    ax[r,j].set_ylim(0,150000000)
plt.show()
```

**출고량이 높은 게임에 대한 분석 및 시각화 프로세스**가 포함되어야 합니다.

```python
#2015년 이후 장르별 출고량 평균 순위

big_game = df.copy()
big_game.index = big_game['Name']

Genre_total_trend = pd.DataFrame(big_game[big_game['Year']>=2015].groupby('Genre')['total'].mean())
Genre_total_trend.reset_index(inplace = True)
Genre_total_trend.sort_values('total', ascending = False, inplace = True)

plt.figure(figsize = (10,6))

sns.barplot(
    x='Genre',
    y='total',
    data=Genre_total_trend
);
```

```python
# 2015년 이후 평균 출고량이 가장 높은 플랫폼

Platform_total_trend = pd.DataFrame(big_game[big_game['Year']>=2015].groupby('Platform')['total'].mean())
Platform_total_trend.reset_index(inplace = True)
Platform_total_trend.sort_values('total', ascending = False, inplace = True)

plt.figure(figsize = (10,6))

sns.barplot(
    x='Platform',
    y='total',
    data=Platform_total_trend
)
```

**2015년 이후 출고량이 높은 장르 Shooter와 출고량이 높은 플랫폼 PS4 게임을 만들자**
